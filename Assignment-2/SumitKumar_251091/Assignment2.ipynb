## Profit v/s Spot Prices graphs for various situations 

```python
import numpy as np
import matplotlib.pyplot as plt

# taking input
k = float(input("Enter strike price : "))
p = float(input("Enter premium paid : "))
a = float(input("Enter lower limit : "))
b = float(input("Enter upper limit : "))

# make a range array of spot prices between the upper limit and lower limit
spot_prices = np.linspace(a, b, 100)
buy_call_profit = []
sell_call_profit = []
buy_put_profit = []
sell_put_profit = []

# calculate profit for each spot price in all conditions
for i in range(100):
    buy_call_profit.append(max(0, spot_prices[i] - k) - p)
    sell_call_profit.append(p - max(0, spot_prices[i] - k))
    buy_put_profit.append(max(0, k - spot_prices[i]) - p)
    sell_put_profit.append(p - max(0, k - spot_prices[i]))
    

# plotting 
plt.plot(spot_prices, buy_call_profit, linewidth = 2)
plt.axhline(y= 0, color = 'black')
plt.axvline(x = k, color = 'gray' , linestyle = '--', label = 'Strike Price')
plt.legend()
plt.title("Buy Call")
plt.xlabel("Spot Price")
plt.ylabel("Profit")
```

##### Inference : The buyer can gain very high profit if the stock price increases drastically but the maximum loss will always be the premium that he paid.

### Selling Call Option

```python
plt.plot(spot_prices, sell_call_profit, linewidth = 2, color = 'red')
plt.axhline(y= 0, color = 'black')
plt.axvline(x = k, color = 'gray' , linestyle = '--', label = 'Strike Price')
plt.legend()
plt.title("Sell Call")
plt.xlabel("Spot Price")
plt.ylabel("Profit")
```

##### Inference : If the stock price rises significantly, the seller will face heavy loss, but if it crashes then he will gain a profit equal to the premium.

### Buying Put Option

```python
plt.plot(spot_prices, buy_put_profit, linewidth = 2, color = 'green')
plt.axhline(y= 0, color = 'black')
plt.axvline(x = k, color = 'gray' , linestyle = '--', label = 'Strike Price')
plt.legend()
plt.title("Buy Put")
plt.xlabel("Spot Price")
plt.ylabel("Profit")
```

##### Inference : The buyer can get high profit if the stock falls and loss is only limited to the amount equal to premium.

### Selling Put Option


```python
plt.plot(spot_prices, sell_put_profit, linewidth = 2, color = 'orange')
plt.axhline(y= 0, color = 'black')
plt.axvline(x = k, color = 'gray' , linestyle = '--',label = 'Strike Price')
plt.legend()
plt.title("Sell Put")
plt.xlabel("Spot Price")
plt.ylabel("Profit")
```

##### Inference : The seller can face heavy loss if the stock crashes, and profit is only limited to premium.
