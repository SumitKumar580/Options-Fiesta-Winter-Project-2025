QUESTION 2

```python
#First the easy method
strs = input()
if strs == strs[::-1]:
  print("palindrome")
else:
  print("No")

#Now, the two pointers
l1 = 0
l2 = len(strs) - 1
while l1 < l2:
  if strs[l1] != strs[l2]:
    print("No")
    break
  l1+=1
  l2-=1
if not l1 < l2:
  print("Palindrome")
```

QUESTION 3

```python
arr = list(map(int, input().split()))
x = int(input())
arr.sort()
l1 = 0
l2 = 1
while l2 < len(arr):
  diff = arr[l2] - arr[l1]
  if diff == x:
    print(arr[l2], arr[l1])
    break
  if diff < x:
    l2+=1
  if diff > x:
    l1+=1
```
